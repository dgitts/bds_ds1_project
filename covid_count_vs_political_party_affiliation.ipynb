```python
import sqlite3
import pandas as pd

# import csvs into dataframes
county_covid_df = pd.read_csv('data/county_covid_data_20200716.csv')
county_votes_df = pd.read_csv('data/us_rep_2018_votes_tx_counties.csv')
fips_df = pd.read_csv('data/county_fips.csv')

# Merge covid df with fips data
county_covid_df = pd.merge(fips_df, county_covid_df, on='county')
county_covid_df.head()
```

```python
# Merge votes df with fips data
county_votes_df = pd.merge(fips_df, county_votes_df, on='county')
county_votes_df.head()
```

```python
# Merge votes and covid dfs
cols_to_use = county_votes_df.columns.difference(county_covid_df.columns)
county_combined_votes_covid_df = pd.merge(county_covid_df, county_votes_df[cols_to_use],left_index=True, right_index=True, how='outer')
county_combined_votes_covid_df.head()
```

```python
county_population_df = pd.read_csv('data/county_population.csv')
county_population_df.head()
```

```python
county_combined_votes_covid_df = pd.merge(county_combined_votes_covid_df, county_population_df, on='county')
county_combined_votes_covid_df.head()
```

```python
from datetime import date

datestamp = date.today().strftime("%Y%m%d")
filename = 'county_combined_covid_vote_data_{}.csv'.format(datestamp)
county_combined_votes_covid_df.to_csv(r'data/' + filename, index = False)
```
